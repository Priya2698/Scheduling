```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
data1 = pd.read_csv("plot_data1.csv")
data2 = pd.read_csv("plot_data2.csv")
```

```python
data1.shape
```

```python
main_data = pd.concat([data1, data2])
```

```python
clipped_data = main_data.loc[(main_data['Time Interval(secs)'] >= 22000) & (main_data['Time Interval(secs)'] <= 55000)]
```

```python
x = clipped_data["Time Interval(secs)"] - 22000
clipped_data["Time Interval(secs)_new"] = x
clipped_data.head()
```

```python
clipped_data.head()
```

```python
plt.figure(figsize=(7,5))
# plt.yscale("log")
# plt.ylim(0,5)

plot=sns.lineplot(x="Time Interval(secs)_new", y="Runtime(in secs)", hue="Configuration", data=clipped_data)
sns.scatterplot(x="Time Interval(secs)_new", y="Runtime(in secs)", hue="Configuration", data=clipped_data,legend=None)

# plt.legend(plot[:2], ['first', 'second']);
# plt.legend(loc='lower center')
plt.legend(["Job1 (running continuously)","Job2 (every 30 minutes)"])
plt.ylim(0,0.3)
plt.xlabel("Wall-clock time (seconds)")
plt.ylabel("Execution time (seconds)")
plt.savefig('myimage.png', format='png', dpi=400)
```
